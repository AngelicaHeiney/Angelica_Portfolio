```python
data.drop_duplicates(keep = False, inplace = True)
```

```python
import matplotlib.pyplot as plt
```

```python
data['Product type'].value_counts().plot(kind='pie')
plt.title('Product Types')
plt.show()
```

```python
best_sellers = data.groupby('Product type')['Number of products sold'].sum().sort_values(ascending=False)
best_sellers
```

```python
data['Revenue Contribution (%)'] = (data['Revenue generated'] / data['Revenue generated'].sum()) * 100
data['Revenue Contribution (%)']
```

```python
plt.figure(figsize=(6, 4))
data.groupby('Product type')['Revenue generated'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('Revenue by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_products = data.groupby('Product type')['Revenue generated'].sum().sort_values(ascending=False).head(3)
print("\nTop 3 revenue-generating products:")
print(top_products)
```

Skincare is the product type with more revenue. Haircare is the product with more Defect Rates

Mumbai is the region with shorter lead times and at the same time with the most revenue by region. Delhi has the longer Lead times and smallest renevue. You should pay attention to this region.

There is no a significant difference between carriers according to shipping costs but Carrier B is the fastest carrier with a mean of 5.3

Route C is the best option if speed is the main concern, despite being the most expensive.
Route A is the best option if cost is the priority, even though it is the slowest.
Route B provides a balanced option with moderate shipping time and cost.


Best Supplier Based on Priorities:
Balanced option: Supplier 2 or Supplier 3
Speed priority: Supplier 3.
Quality priority: Supplier 1.
Cost priority: Supplier 2.

If you want the best overall balance, choose Supplier 2 (low cost, moderate defect rate, moderate lead time).
If lead time is the most critical, choose Supplier 3 (fastest delivery).
If quality is paramount, choose Supplier 1 (lowest defect rate).

Transportation modes:
Road the fastest but the second most expensive next to Air
Sea is the slowest but the chepeast.
Order of efficiency (fast to slow):
Road > Air > Rail > Sea

For overall balance:
Fast delivery priority: Use Road (fastest shipping time with moderate cost).
Low cost priority: Use Sea (cheapest shipping option with the lowest cost per day).
Balanced option: Use Rail (moderate shipping time and cost).
